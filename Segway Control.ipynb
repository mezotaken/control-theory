$
\text{Схематическое изображение сегвея, представляющего собой перевернутый маятник,}\\
\text{закреплённый на оси вращения колеса, которое приводится в движение электродвигателем}\\
$
    



$ 
F_{тяг} = K_t(u - K_s\dot{x})~- \text{тяговое усилие, которое обеспечивает электродвигатель.} \\ 
u ~-\text{напряжение, подаваемое на обмотку электродвигателя.} \\
$

$
\text{Параметры системы:}\\
K_t = 1.726~ \frac{Н}{В} \\ 
K_s = 4.487~ \frac{В·с}{м} \\
m = 0.56~ кг~- \text{масса маятника.} \\
M = 1.206~ кг~- \text{масса тележки.} \\
l = 0.177~ м~- \text{расстояние от точки крепления до центра масс.} \\
I = 0.89~ кг·м^2 ~- \text{момент инерции маятника относительно центра масс.} \\
B_{eq} = 5.4~ \frac{Н·с}{м}~- \text{коэффициент вязкого трения между колесом каретки и направляющей.} \\
B_p = 1.4~ Н·м·с ~- \text{коэффициент вязкого трения в точке крепления маятника.} \\
\\
\\
\text{Переменные состояния:}~~~  \xi = (\theta , x , \dot{\theta}, \dot{x}) \\
$

### 1. Кинетическая энергия системы  
<font size="3">
$
~\\
\text{Кинетическая энергия = энергия поступательного движения тележки}\\
\text{                      + энергия вращательного движения маятника}\\
\text{                        вокруг своего центра масс}\\
\text{                      + кинетическая энергия движения центра масс маятника}\\
T = \cfrac{M \dot{x}^2 }{2} + \cfrac{I \dot{\theta}^2 }{2} + \cfrac{m \Big[(\cfrac{d}{dt}(x-l\sin{\theta}))^2 + (\cfrac{d}{dt}(l\cos{\theta}))^2\Big] }{2} \\
= \cfrac{M \dot{x}^2 }{2} + \cfrac{I \dot{\theta}^2 }{2} + \cfrac{m [(\dot{x}-l\dot{\theta}\cos{\theta})^2 + (l\dot{\theta}\sin{\theta})^2] }{2} \\
= \cfrac{M \dot{x}^2 }{2} + \cfrac{I \dot{\theta}^2 }{2} + \cfrac{m(\dot{x}^2 - 2l\dot{x}\dot{\theta}\cos{\theta} + l^2\dot{\theta}^2)}{2}\\
= \cfrac{(M+m) \dot{x}^2 }{2} + \cfrac{(I+ ml^2) \dot{\theta}^2 }{2} -ml\dot{x}\dot{\theta}\cos{\theta} \\
$

### 2. Потенциальная энергия системы
<font size="3">
$
~\\
W = mgl\cos{\theta}
$

### 3. Уравнения, описывающие динамику системы
<font size="3">
$
~\\
L = \cfrac{(M+m) \dot{x}^2 }{2} + \cfrac{(I+ ml^2) \dot{\theta}^2 }{2} -ml\dot{x}\dot{\theta}\cos{\theta} - mgl\cos{\theta}\\
\begin{cases}
    \cfrac{d}{dt}\Big(\cfrac{\partial{L}}{\partial{\dot{x}}}\Big) -\cfrac{\partial{L}}{\partial{x}} = F_{тяг} - B_{eq}\dot{x}       \\
    \cfrac{d}{dt}\Big(\cfrac{\partial{L}}{\partial{\dot{\theta}}}\Big) -\cfrac{\partial{L}}{\partial{\theta}} = -                   B_p\dot{\theta}
\end{cases} \\
\begin{cases}
    \cfrac{d}{dt}\Big( (M+m)\dot{x} -ml\dot{\theta}\cos{\theta} \Big)  = F_{тяг} - B_{eq}\dot{x} \\
    \cfrac{d}{dt}\Big( (I+ml^2)\dot{\theta} - ml\dot{x}\cos{\theta} \Big) - ml\dot{x}\dot{\theta}\sin{\theta} - mgl\sin{\theta}     = - B_p\dot{\theta}
\end{cases} \\
\begin{cases}
    (M+m)\ddot{x} - ml\ddot{\theta}\cos{\theta} + ml\dot{\theta}^2\sin{\theta}  = F_{тяг} - B_{eq}\dot{x} \\
    (I+ml^2)\ddot{\theta} - ml\ddot{x}\cos{\theta}  - mgl\sin{\theta}     = - B_p\dot{\theta}
\end{cases} \\
$

### 4. Нахождение с.р. и линеаризация в окрестности неустойчивого с.р.
<font size="3">
$
~\\
\text{СЛУ относительно } \ddot{x} \: и \: \ddot{\theta}, \text{применим метод Крамера:} \\
\begin{cases}
    \dot{\theta} = \dot{\theta}\\
    \dot{x} = \dot{x}\\
    \ddot{\theta} = \cfrac{(K_t(u - K_s\dot{x}) - B_{eq}\dot{x} - ml\dot{\theta}^2\sin{\theta})(ml\cos{\theta}) + (M+m)(-B_p \dot{\theta} + mgl\sin{\theta})}{(I+ml^2)(M+m) - m^2 l^2\cos^2{\theta}} \\
    \ddot{x} = \cfrac{(ml\cos{\theta})(-B_p \dot{\theta} + mgl\sin{\theta}) + (I+ml^2)(K_t(u - K_s\dot{x}) - B_{eq}\dot{x} - ml\dot{\theta}^2\sin{\theta})}{(I+ml^2)(M+m) - m^2 l^2\cos^2{\theta}}
\end{cases} \\
\text{Неустойчивое с.р.:}~~(0,x_0,0,0)^T \\
\text{Устойчивое с.р.:}~~(\pi,x_0,0,0)^T \\
\text{Линеаризованная в окрестности неустойчивого с.р. система:}\\
\begin{cases}
\dot{\theta} = \dot{\theta}\\
\dot{x} = \dot{x}\\
\ddot{\theta} = \cfrac{(M+m)mgl\theta - B_p(M+m)\dot{\theta} - ml(K_t K_s + B_{eq})\dot{x} + mlK_t u}{(I+ml^2)(M+m) - m^2 l^2} \\
\ddot{x} = \cfrac{m^2 l^2 g\theta - mlB_p\dot{\theta} - (I+ml^2)(K_t K_s + B_{eq})\dot{x} + (I+ml^2)K_t u}{(I+ml^2)(M+m) - m^2 l^2}
\end{cases}
$
